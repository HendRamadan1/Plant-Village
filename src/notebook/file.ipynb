```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
probe_df = pd.read_csv(r'D:\hackathon\data_HERE_cairo_hackathon_2024\probe_data\0\probe_2024_07_08_00_00_00.csv') 


geometry = [Point(xy) for xy in zip(probe_df['longitude'], probe_df['latitude'])]
probe_gdf = gpd.GeoDataFrame(probe_df, geometry=geometry, crs="EPSG:4326")


roundabouts_df = pd.read_csv(r'D:\hackathon\data_HERE_cairo_hackathon_2024\hamburg_extra_layers\hamburg_rounsabouts.csv')
geometry_rb = [Point(xy) for xy in zip(roundabouts_df['longitude'], roundabouts_df['latitude'])]
roundabouts_gdf = gpd.GeoDataFrame(roundabouts_df, geometry=geometry_rb, crs="EPSG:4326")

# 3. الرسم (Visualization)
fig, ax = plt.subplots(figsize=(12, 8))

# رسم نقط العربيات باللون الرمادي الشفاف
probe_gdf.plot(ax=ax, color='gray', markersize=1, alpha=0.5, label='Car Movements (Probe)')

# رسم مراكز الدورانات باللون الأحمر الواضح
roundabouts_gdf.plot(ax=ax, color='red', markersize=50, marker='o', label='Actual Roundabouts')

plt.title("Visualizing Probe Data vs. Known Roundabouts in Hamburg")
plt.legend()
plt.show()
```
